```python
# !pip install seaborn 
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
import numpy as np 
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.columns = ['total_bill', 'tip', 'gender', 'smoker', 'day', 'time', 'member']
```

```python
# sns.lineplot()
# sns.boxplot()
# sns.scatterplot() 
# sns.countplot() 
# sns.barplot()
# sns.histplot()
sns.scatterplot(x='total_bill',y='tip',data=df) 
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,kind='line')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='gender',style='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='gender',style='smoker',col='time')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='gender',col='day')
plt.show()
```
